```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('dataset_1.csv')
```

```python
data.head()
```

```python
data.drop('user', inplace=True, axis=1)
data.drop('fall_accident', inplace=True, axis=1)
```

```python
import random
# Changing blood pressure to add systole and diastole
systole = []
for i in range(10000):
    tmp = random.randint(100, 170)
    systole.append(tmp)
# Replacing heart rate
data['systole'] = systole
```

```python
data.head()
```

```python

# Changing blood pressure to add systole and diastole
diastole = []
for i in range(10000):
    tmp = random.randint(60, 90)
    diastole.append(tmp)
# Replacing heart rate
data['diastole'] = diastole
```

```python
data.head()
```

```python
#data.drop('gyroX', inplace=True, axis=1)
data.drop('gyroY:', inplace=True, axis=1)
data.drop('gyroZ', inplace=True, axis=1)
```

```python
data.drop("body_temp", inplace=True, axis=1)
data.drop("mgr", inplace=True, axis=1)
data.drop("cardiac_arrest", inplace=True, axis=1)
```

```python
data.head()
```

```python
data.drop("pressure", inplace=True, axis=1)
```

```python
# Figuring out if they have high blood pressure
risk_list = []
for i in range(10000):
    sys = data.iloc[i:i+1, 2:3].values
    dia = data.iloc[i:i+1, 3:4].values
    print(sys[0][0])
    if (sys[0][0] > 125 or dia[0][0] > 80):
        risk = 1
    else:
        risk = 0
    risk_list.append(risk)
data["risk"] = risk_list
```

```python
data.to_csv("final_data.csv")
```
