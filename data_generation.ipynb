```python
# Generating walking asymmetry

walkingAsymmetry = []

# My current metric is that if the gyroY and gyroZ differs by more than 0.75, I classify it as walking asymmetry
for i in range(10000):
    difference = abs(data.loc[i:i+1, "gyroY:"].any().astype(int) - data.loc[i:i+1, "gyroZ"].any().astype(int))
    if (difference > 0.75):
        walkingAsymmetry.append(1)
    else:
        walkingAsymmetry.append(0)
data["walkingAsymmetry"] = walkingAsymmetry
```
